```python
df['sport'] = df.apply(lambda row: int(any('sport' in str(x).lower() for x in row)), axis=1)
df['education'] = df.apply(lambda row: int(any('education' in str(x).lower() for x in row)), axis=1)
df['free_play'] = df.apply(lambda row: int(any('Free to Play' in str(x).lower() for x in row)), axis=1)
df['animation_modeling'] = df.apply(lambda row: int(any('Animation & Modeling' in str(x).lower() for x in row)), axis=1)
df['action'] = df.apply(lambda row: int(any('Action' in str(x).lower() for x in row)), axis=1)
df['simulation'] = df.apply(lambda row: int(any('Simulation' in str(x).lower() for x in row)), axis=1)
df['racing'] = df.apply(lambda row: int(any('Racing' in str(x).lower() for x in row)), axis=1)
df['video_production'] = df.apply(lambda row: int(any('Video Production' in str(x).lower() for x in row)), axis=1)
```

```python
df['original_price'] = df['original_price'].str.replace('₹', '')
df['original_price'] = df['original_price'].str.replace(',', '')
df['original_price'] = df['original_price'].astype(float)
```

```python
content_descriptor = df["content_descriptor"].unique()
content_descriptor
```

```python
df['violence'] = df.apply(lambda row: int(any('violence' in str(x).lower() for x in row)), axis=1)
df['sex'] = df.apply(lambda row: int(any('sex' in str(x).lower() for x in row)), axis=1)
df['drugs'] = df.apply(lambda row: int(any('drugs' in str(x).lower() for x in row)), axis=1)
df['bad'] = df.apply(lambda row: int(any('sport' in str(x).lower() for x in row)), axis=1)
df['blood'] = df.apply(lambda row: int(any('blood' in str(x).lower() for x in row)), axis=1)
```

```python
df
```

```python
X = df[["sport", "education", "free_play", "animation_modeling", "action", "simulation", "racing", "video_production", "violence", "sex", "drugs", "bad", "blood"]]
Y = df["age_rating"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

# MODELES

1) Linear Regression

```python
model_lr = LinearRegression()
```

```python
model_lr.fit(X_train, Y_train)
```

```python
print('Coefficients:', model_lr.coef_)
```

```python
print('Intercept:', model_lr.intercept_)
```

```python
Y_pred = model_lr.predict(X_test)
```

```python
mse_lr = mean_squared_error(Y_test, Y_pred)
mse_lr
```

```python
r2 = r2_score(Y_test, Y_pred)
r2
```

```python
with open('video_game_lr_model.pkl', 'wb') as file:
    pickle.dump(model_lr, file)
```

2) Sequential

```python
model_seq = Sequential()
```

```python
model_seq.add(Dense(10, input_dim=X_train.shape[1], activation='relu'))
model_seq.add(Dense(1, activation='linear'))
```

```python
model_seq.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse'])
```

```python
history = model_seq.fit(X_train, Y_train, epochs=15, batch_size=32, validation_data=(X_test, Y_test))
```

```python
loss_seq, mse_seq = model_seq.evaluate(X_test, Y_test)
```

```python
loss_seq
```

```python
mse_seq
```

```python
with open('video_game_seq_model.pkl', 'wb') as file:
    pickle.dump(model_seq, file)
```

# COMPARAISON MODELES

Performance de la régression linéaire
- Mean Squared Error (MSE) : 0.0123
- R-squared (R²) : 0.78

Performance du modèle séquentiel (réseau de neurones)
- Mean Squared Error (MSE) : 0.0122

# CHOIX DU MODELE 

En termes de MSE, les deux modèles semblent avoir des performances assez similaires sur votre ensemble de données de test. 

Mais on choisie le modèle Linear Regression car le Sequential est une boite noire et plus complexe à expliquer
